MY MAP 「麗澤大学付近の神社」

通学中に神社をよく見るので、まとめてみました。
留学生の方にもおすすめのマップができたと思います。
あまり深くは調べてませんが、歴史があり、地元の方によってきれいな状態で大切に保たれていることがわかったので今度訪れてみたいと思いました。

pop up は日本語だと縦書きになるのでしょうか？？

```python
# import folium, a leaflet python library
import folium

# create a list of lat/lon pair for UCLA
reitaku = [35.8330, 139.9550]
itsukushima = [35.8346267642589, 139.94902440400884]
nakashinjukufujisengen = [35.83523562025213, 139.94906731915918]
inari = [35.82790094045206, 139.94453393859212]
katori = [35.83880099972304, 139.9465939230242]

# make the map
m = folium.Map(location=reitaku, zoom_start=15)

# add a marker
folium.Marker(reitaku, popup="Reitaku University, the best place on earth!", tooltip="Reitaku University").add_to(m)
folium.Marker(itsukushima, popup="森の中で自然を感じられる神社", tooltip="厳島神社").add_to(m)
folium.Marker(nakashinjukufujisengen, popup="500年の歴史が?!", tooltip="中新宿富士浅間神社").add_to(m)
folium.Marker(inari, popup="珍しいかやぶき屋根の神社", tooltip="稲荷神社").add_to(m)
folium.Marker(katori, popup="開けた場所の明るい神社", tooltip="香取神社").add_to(m)

# show the map
m
```
